```python
import matplotlib.pyplot as plt
import numpy as np

fac=2

# Specify the Palatino font family and font name
font = {'family': 'serif',
        'serif': ['Palatino'],
        'size': 8*fac}  # You can adjust the font size as needed

plt.rc('font', **font)
plt.rc('text', usetex=True)  # Enable LaTeX rendering for text

width_cm=9*fac
height_cm=7*fac
```

```python
# Load data from the first file (assuming it's a text file with 4 columns)
# File contains the polynomial expansion 
file1_data = np.loadtxt('plots/scalar_sigmav_all.out')
x = file1_data[:, 0]
y1 = file1_data[:, 1] #formula
y2 = file1_data[:, 2] #low-T expanzion
y3 = file1_data[:, 3] #Prediction

xmin=x.min()
xmax=x.max()

# Create a new figure
plt.figure(figsize=(width_cm / 2.54, height_cm / 2.54))  # Convert cm to inches (1 inch = 2.54 cm)

# Plot the first curve with a double log scale
plt.semilogx(x, y2, label='Taylor expansion', color='red')
plt.semilogy(x, y1, label='Numerical expression', color='blue')
plt.semilogy(x, y3, label='Analytical from literature', color='green',linestyle='--')

# Set LaTeX labels for the axes
plt.xlabel(r'$T/m_\chi$', fontsize=12)
plt.ylabel(r'$\langle \sigma v \rangle (\mathrm{GeV}^{-2})$', fontsize=12)

# Create a legend with transparent background, on the top left, and without a border
plt.legend(loc='upper left', frameon=False, bbox_to_anchor=(0, 1))
# Customize the tick parameters
plt.tick_params(axis='both', which='both', direction='in', top=True, right=True, bottom=True, left=True)

#plt.xlim(xmin, xmax)
#plt.ylim(1.0e-36, ymax)

# Show the plot
plt.grid(False)

plt.savefig('scalar_sigmav.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

```python
# Load data from the first file (assuming it's a text file with 4 columns)
# File contains the polynomial expansion 
file1_data = np.loadtxt('plots/Weff.out')
x = file1_data[:, 0]
y1 = file1_data[:, 3] #prediction
y2 = file1_data[:, 4] #numerical

# Create a new figure
plt.figure(figsize=(width_cm / 2.54, height_cm / 2.54))  # Convert cm to inches (1 inch = 2.54 cm)

# Plot the first curve with a double log scale
plt.semilogx(x, y2, label='Numerical', color='red')
plt.semilogy(x, y1, label='Prediction', color='blue')

# Set LaTeX labels for the axes
plt.xlabel(r'$E_{cm}$(GeV)', fontsize=12)
plt.ylabel(r'$W_{eff}$', fontsize=12)

# Create a legend with transparent background, on the top right, and without a border
plt.legend(loc='upper right', frameon=False, bbox_to_anchor=(1, 1))
# Customize the tick parameters
plt.tick_params(axis='both', which='both', direction='in', top=True, right=True, bottom=True, left=True)

#plt.xlim(xmin, xmax)
plt.ylim(1.0e-2, 5.0)

# Show the plot
plt.grid(False)

plt.savefig('scalar_weff.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
